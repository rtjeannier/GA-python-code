##### Problem 5: Build a function that will take in lists Y and X, where Y and X both contain quantitative data, and returns the estimates of $\beta_0$ and $\beta_1$ that minimize SSE. Test your function on price and the variable you determined was best in Problem 4. Report the slope and Y-intercept.

```python
import scipy as sp

def minimize_SSE(X, Y):
    # Calculate sample means of X, Y
    xbar = np.mean(X)
    ybar = np.mean(Y)
    # Calculate STD of X, Y
    x_std = np.std(X,ddof=1)
    y_std = np.std(Y,ddof=1)
    # Pearson Correlation
    r_xy = sp.stats.pearsonr(X,Y)
    
    B1 = r_xy[0] * y_std / x_std
    B0 = ybar - B1 * xbar

    return (B0, B1)

        
```

```python
minimize_SSE(df_clean.sq__ft, df_clean.price)
```

The y intercept is 21519 and the slop is 133.

##### Problem 6: Interpret the Y-intercept. Does it make sense to interpret the Y-intercept here?

The Y intercept represents the price of a 0 sqr_ft house. A 0 sqr_ft house does not exist, and we earlier removed it from out dataset. So, from a prediction standpoint, it is not a valid interpretation. What we can infer from a Y intercept is that our model show a relationship between price and sqr_ft that is offset from zero and rises positively. In otherwords for all reasonable ranges of sqr_ft price will be a postive number. This is good. Had we had a negative Y intercept, we'd have to see at which sq_ft our price went negative and determine if it was within a range we'd expect at the very least a positive price.
